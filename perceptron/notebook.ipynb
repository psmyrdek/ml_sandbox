```python
import random
import matplotlib.pyplot as plt
from data import or_data

epoch_limit = 1000
learning_rate = 0.01
weights = [random.uniform(-0.1, 0.5), random.uniform(-0.1, 0.5)]
threshold = [random.uniform(-0.5, 0.5)]

weights_history = []
threshold_history = []
epoch_errors = []

print(weights)
print(threshold)
```

```python
def calc_dot_product(input_vector, weights, threshold):
    return (input_vector[0] * weights[0] + input_vector[1] * weights[1]) + threshold[0]


def activate(dot_product):
    return 1 if dot_product >= 0 else 0

def train_perceptron(training_data, weights, threshold, learning_rate):
    cur_epoch = 0
    trained = False

    while cur_epoch < epoch_limit and not trained:
        has_error = False
        total_error = 0
        for input_vector, target in training_data:
            dot_product = calc_dot_product(input_vector, weights, threshold)
            output = activate(dot_product)
            error = target - output
            total_error += abs(error)
            print(f"Target: {target}, Output: {output}")

            if error != 0:
                has_error = True
                weights[0] += input_vector[0] * learning_rate * error
                weights[1] += input_vector[1] * learning_rate * error
                # Uncomment this line if you want to allow threshold updates
                threshold[0] += learning_rate * error

        # Store weights and threshold for each epoch
        weights_history.append(weights[:]) # Append a copy of the weights
        threshold_history.append(threshold[0])
        epoch_errors.append(total_error)

        if not has_error:
            trained = True
        cur_epoch += 1

    print("Training completed in epochs:", cur_epoch)

train_perceptron(or_data, weights, threshold, learning_rate)
```

```python
# After training
epochs = range(len(weights_history))

# Plot weights and threshold history
plt.figure(figsize=(12, 5))

# Plot weights and threshold over epochs
plt.subplot(1, 2, 1)
plt.plot(epochs, [w[0] for w in weights_history], label="Weight 1")
plt.plot(epochs, [w[1] for w in weights_history], label="Weight 2")
plt.plot(epochs, threshold_history, label="Threshold")
plt.xlabel("Epochs")
plt.ylabel("Value")
plt.title("Weights and Threshold Over Epochs")
plt.legend()

# Plot total error over epochs
plt.subplot(1, 2, 2)
plt.plot(epochs, epoch_errors, label="Total Error", color='r')
plt.xlabel("Epochs")
plt.ylabel("Total Error")
plt.title("Total Error Over Epochs")
plt.legend()

plt.tight_layout()
plt.show()
```
